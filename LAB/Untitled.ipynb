# LAB 2 Write a Python program to insert an element into a sorted list

```python
my_list = [2, 4, 1, 5, 6]
my_list.sort()

ch = 'y'

while ch == 'y':
    ele = int(input("Enter an element: "))
    size = len(my_list)  # Update size after every change

    for i in range(size):
        if ele < my_list[i]:
            my_list.insert(i, ele)
            break
    else:
        # if not inserted in the loop, it's the largest element
        my_list.append(ele)

    print("Updated list:", my_list)
    ch = input("Want to continue? (y/n): ")
```

# LAB 3 Write a python program using object oriented programming 
# to demonstrate encapsulation , overloading with inheritance


```python
class Stud:
    def __init__(self):
        self.name = "ABC"
        self.marks = 0

    def getDetails(self):
        self.name = input("Enter Name ")
        self.marks = input("Enter Marks ")

    def display(self,name=None,marks=None):
        if name is not None:
            print(f"Name is {name}")
        else:
            print(f"Name is {self.name}")

        if marks is not None:
            print(f"Marks is {marks}")
        else:
            print(f"Marks is {self.marks}")


class PgStud(Stud):
    def __init__(self):
        super().__init__()
        self.specialization = ""

    def getDetails(self):
        super().getDetails()
        self.specialization = input("Enter specialization ")

    def display(self,name=None,marks=None):
        super().display()
        print(f"Specialization is {self.specialization}")

        
 
print("Encapsulation & Inheritance Demo")
pg = PgStud()
pg.getDetails()

print("\nFunction Overloading Demo (display with 0, 1, 2 arguments):")
pg.display()                          # 0 arguments
pg.display("Alice")                   # 1 argument
pg.display("Bob", 90)                 # 2 arguments

print("\nMethod Overriding Demo:")
pg.display()  # Calls overridden method in PGStudent
```

```python
# =====================
# Array Manipulation
# =====================

# Reshape
x = np.array([[2, 3, 4], [5, 6, 7]])
print("Original array:\n", x)

x = np.reshape(x, (3, 2))
print("\nReshaped to 3x2:\n", x)
```

```python
# Flat
x = np.array([[2, 3, 4], [5, 6, 7], [12, 13, 14]])
print("\nOriginal array for flat:\n", x)

x.flat[[2, 5, 7]] = 155  # Set specific flat indices
print("\nAfter setting flat indices to 155:\n", x)

x.flat = 10  # Set all elements to 10
print("\nAfter setting all elements to 10:\n", x)
```

```python
# Transposing
A = np.arange(10).reshape(2,5)
print('\n',A)

A = np.arange(10).reshape(2, 5)
print("\nOriginal A:\n", A)

B = A.T
print("\nTransposed A:")
print(B)
```

```python
# =====================
# Broadcasting
# =====================
a = np.array([[1, 2, 3, 4], [2, 4, 5, 6], [10, 20, 39, 3]])
b = np.array([2, 4, 6, 8])

print("\nMatrix A:\n", a)
print("\nArray B:\n", b)

print("\nA + B:\n", a + b)
print("\nA - B:\n", a - b)
print("\nA * B:\n", a * b)
print("\nA / B:\n", a / b)
```

# LAB 7: Write a Python program that creates a mxn integer arrayand Prints its attributes using  matplotlib 

```python
import pandas as pd
import random
import matplotlib.pyplot as plt

# Get value for m (number of data points)
m = int(input("Enter the number of records (m): "))

# Create an empty DataFrame
df = pd.DataFrame()

# Random data generation
age = random.sample(range(17, 61), m)  # random ages between 1760
height = [round(random.uniform(4.0, 6.5), 2) for _ in range(m)]  # height in ft
weight = random.sample(range(40, 90), m)  # weight in kg

# Add columns to DataFrame
df["age"] = age
df["height"] = height
df["weight"] = weight

# Sort data by age for better plotting
df.sort_values(by="age", inplace=True)
print("\nGenerated DataFrame:")
print(df)

# Plotting
plt.plot(df["age"], df["height"], color="red", marker="*", label="Height")
plt.plot(df["age"], df["weight"], color="green", marker="o", label="Weight")
plt.xlabel("Age")
plt.ylabel("Height (ft) / Weight (kg)")
plt.title("Height and Weight vs Age")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# LAB 9  Write a Python program to demonstrate the generation of logistic regression models using Python

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression

# Step 1: Generate synthetic regression data
x1, y1 = datasets.make_regression(n_samples=1000, n_features=1, noise=100)

# Step 2: Print the generated data (optional)
print("X values (features):\n", x1[:5])
print("Y values (target):\n", y1[:5])

# Step 3: Scatter plot the original data points
plt.scatter(x1, y1, marker="*", color="green", label="Actual Data")

# Step 4: Create a Linear Regression model
m = LinearRegression()

# Step 5: Train the model with the data
m.fit(x1, y1)

# Step 6: Use the trained model to make predictions
yp = m.predict(x1)

# Step 7: Plot the predicted regression line
plt.plot(x1, yp, 'r--', label="Regression Line")

# Step 8: Add labels and title
plt.title("Linear Regression")
plt.xlabel("Feature (x1)")
plt.ylabel("Target (y1)")
plt.legend()
plt.grid(True)
plt.show()
```

# LAB 9: Write a Python program to demonstrate the generation of logistic regression models using Python.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

# Load the dataset (CSV should have 'age' and 'insurance' columns)
df = pd.read_csv("data sets/ini.csv")

# Scatter plot of raw data
plt.scatter(df.age, df.insurance, color="blue", marker="*")
plt.xlabel("Age of the person")
plt.ylabel("Policy claim")
plt.title("Logistic Regression")
plt.show()

# Split the dataset (60% train, 40% test)
xt, xtest, yt, ytest = train_test_split(df[['age']], df.insurance, test_size=0.4)

# Create Logistic Regression model
lg = LogisticRegression()
lg.fit(xt, yt)

# Predict using the test set
predictions = lg.predict(xtest)

# Print test values and predictions
print("\nTest Features:\n", xtest.head())
print("\nPredictions:\n", predictions[:5])

# Evaluate model
print("\nConfusion Matrix:\n", confusion_matrix(ytest, predictions))
print("\nAccuracy Score:", accuracy_score(ytest, predictions))
```

# LAB 11 Write a Python program to demonstrate Data Visualization using Seaborn

```python
# Program 11: Data Visualization using Seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load sample dataset
df = sns.load_dataset('tips')
print(df)

# -------------------- BAR PLOT --------------------
sns.barplot(x='sex', y='total_bill', data=df, palette='plasma')
plt.title("Bar Plot: Total Bill by Sex")
plt.show()

# -------------------- BOX PLOT --------------------
df["weekend"] = df["day"].isin(["Sat", "Sun"])
sns.boxplot(x="day", y="total_bill", hue="weekend", data=df)
plt.title("Box Plot: Total Bill by Day (Weekend vs Weekday)")
plt.show()

# -------------------- VIOLIN PLOT --------------------
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', split=True)
plt.title("Violin Plot: Total Bill by Day and Sex")
plt.show()

# -------------------- STRIP PLOT --------------------
sns.stripplot(x='day', y='total_bill', data=df, jitter=True, hue='smoker', dodge=True)
plt.title("Strip Plot: Total Bill by Day and Smoker")
plt.show()

# -------------------- SWARM PLOT --------------------
sns.swarmplot(x='day', y='total_bill', data=df, palette='plasma')
plt.title("Swarm Plot: Total Bill by Day")
plt.show()

# -------------------- HEATMAP --------------------
stats_df = df.drop(['sex', 'day', 'smoker', 'time'], axis=1)
corr = stats_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlations")
plt.show()

# -------------------- JOINT PLOT --------------------
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
plt.suptitle("Joint Plot: Total Bill vs Tip", y=1.02)
plt.show()

# -------------------- POINT PLOT --------------------
sns.pointplot(x="day", y="size", hue="sex", linestyles=["-", "--"], data=df)
plt.title("Point Plot: Party Size by Day and Sex")
plt.show()

# -------------------- FACET GRID --------------------
df['tip_pct'] = 100 * df['tip'] / df['total_bill']
grid = sns.FacetGrid(df, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15))
plt.subplots_adjust(top=0.9)
grid.fig.suptitle("FacetGrid: Tip % Distribution by Sex and Time")
plt.show()
```
